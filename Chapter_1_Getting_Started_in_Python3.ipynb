## 7. Control Structures

Programs make **decisions** (selection) and **repeat** actions (iteration).


**Listing 1-14 — Simple if:**


````python
# Listing 1-14: Simple if Statement
var = -1
if var < 0:  # Run the indented block only if this condition is True
    print(var)
    print("the value of var is negative")
````

````python
if var > 0:  # Run the indented block only if this condition is True
    print(var)
    print("the value of var is positive")
````

**Listing 1-15 — if-else:**


````python
# Listing 1-15: if-else Statement
var = 1
if var < 0:  # Run the indented block only if this condition is True
    print("the value of var is negative")
    print(var)
else:
    print("the value of var is positive")
    print(var)
````

**Listing 1-16 — Nested if / elif (grade example):**


````python
# Listing 1-16: Nested if-else (grading)
score = 95
if score >= 99:  # Run the indented block only if this condition is True
    print('A')
elif score >= 75:  # Otherwise, check this next condition
    print('B')
elif score >= 60:  # Otherwise, check this next condition
    print('C')
elif score >= 35:  # Otherwise, check this next condition
    print('D')
else:
    print('F')
````

**Listing 1-17 — for loops:**


````python
# Listing 1-17: for Loop Examples
print("First Example")
for item in [1, 2, 3, 4, 5]:  # Repeat the indented block for each item in the sequence
    print('item :', item)
````

````python
print("Second Example")
letters = ['A', 'B', 'C']
for letter in letters:  # Repeat the indented block for each item in the sequence
    print('First loop letter :', letter)
````

````python
print("Third Example - Iterating by sequence index")
for index in range(len(letters)):  # Repeat the indented block for each item in the sequence
    print('First loop letter :', letters[index])
````

````python
print("Fourth Example - Using else with for")
for item in [1, 2, 3, 4, 5]:  # Repeat the indented block for each item in the sequence
    print('item :', item)
else:
    print('looping over item complete!')
````

**Listing 1-18 — while loop:**


````python
# Listing 1-18: while Loop
count = 0
while count < 5:  # Repeat the indented block while this condition stays True
    print('The count is:', count)
    count = count + 1
````

**Listing 1-19 — while with else:**


````python
# Listing 1-19: while with else
while count < 5:  # Repeat the indented block while this condition stays True
    print(count, " is less than 5")
    count = count + 1

else:
    print(count, " is not less than 5")
````

## 15. Exception Handling

An **exception** is an error that interrupts normal program flow. Good programs handle expected and unexpected errors.

### Common Built-in Exceptions (Table 1-17 — selected)

| Exception | When it happens |
|-----------|-----------------|
| `ZeroDivisionError` | Dividing by zero |
| `TypeError` | Wrong type for an operation |
| `ValueError` | Valid type but invalid value |
| `NameError` | Variable name not defined |
| `IndexError` | List index out of range |
| `KeyError` | Dictionary key not found |
| `IOError` | Input/output failure (file issues) |
| `SyntaxError` | Invalid Python syntax |
| `IndentationError` | Incorrect indentation |

### try / except / finally

- **try** — code that might fail
- **except** — handle specific errors
- **finally** — always runs (cleanup)

**Listing 1-52 — Basic try/except:**


````python
# Listing 1-52: Exception Handling
import sys  # Import sys for system-level features (e.g. errors)

try:  # Try to run code that might raise an error
    a = 1
    b = 1
    print("Result of a/b:", a / b)

except ZeroDivisionError:  # Handle a specific type of error if one occurs
    print("Can't divide by zero")

except TypeError:  # Handle a specific type of error if one occurs
    print("Wrong data type, division allowed on numeric types only")

except:  # Handle any other unexpected error
    print("Unexpected error occurred")
````

**Listing 1-53 — Exception handling with file I/O and finally:**


````python
# Listing 1-53: Exception Handling with File Operations
import sys  # Import sys for system-level features (e.g. errors)

f = None
try:  # Try to run code that might raise an error
    f = open('vehicles.txt', 'r')
    s = f.readline()  # Read one line from the file
    print(s)
    i = int(s.strip())

except IOError as e:  # Handle a specific type of error if one occurs
    print("I/O error({0}): {1}".format(e.errno, e.strerror))

except ValueError:  # Handle a specific type of error if one occurs
    print("Could not convert data to an integer.")

except:  # Handle any other unexpected error
    print("Unexpected error occurred")

finally:  # Always run this cleanup code (even after errors)
    if f is not None:  # Run the indented block only if this condition is True
        f.close()  # Close the open file
        print("file has been closed")
````
