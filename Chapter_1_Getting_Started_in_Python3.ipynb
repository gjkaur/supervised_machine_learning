### 5.1 Python Identifiers

**Identifiers** are names for variables, functions, and classes.

**Rules:**
- Letters (a–z, A–Z), digits (0–9), and underscore `_`
- **Cannot** start with a digit (`1var` is invalid, `var1` is valid)
- **Cannot** use reserved **keywords** (see table below)
- No special symbols like `@`, `#`, `$` in names


### 5.2 Python Keywords (Reserved Words)

These words have special meaning — you cannot use them as variable names:

`False`, `None`, `True`, `and`, `as`, `assert`, `async`, `await`, `break`, `class`, `continue`, `def`, `del`, `elif`, `else`, `except`, `finally`, `for`, `from`, `global`, `if`, `import`, `in`, `is`, `lambda`, `nonlocal`, `not`, `or`, `pass`, `raise`, `return`, `try`, `while`, `with`, `yield`

**Listing 1-16 — Nested if / elif (grade example):**


````python
# Listing 1-16: Nested if-else (grading)
score = 95
if score >= 99:
    print('A')
elif score >= 75:
    print('B')
elif score >= 60:
    print('C')
elif score >= 35:
    print('D')
else:
    print('F')
````

**Listing 1-17 — for loops:**


````python
# Listing 1-17: for Loop Examples
print("First Example")
for item in [1, 2, 3, 4, 5]:
    print('item :', item)
````

````python
print("Second Example")
letters = ['A', 'B', 'C']
for letter in letters:
    print('First loop letter :', letter)
````

````python
print("Third Example - Iterating by sequence index")
for index in range(len(letters)):
    print('First loop letter :', letters[index])
````

````python
print("Fourth Example - Using else with for")
for item in [1, 2, 3, 4, 5]:
    print('item :', item)
else:
    print('looping over item complete!')
````

**Listing 1-18 — while loop:**


````python
# Listing 1-18: while Loop
count = 0
while count < 5:
    print('The count is:', count)
    count = count + 1
````

**Listing 1-19 — while with else:**


````python
# Listing 1-19: while with else
while count < 5:
    print(count, " is less than 5")
    count = count + 1

else:
    print(count, " is not less than 5")
````

## 15. Exception Handling

An **exception** is an error that interrupts normal program flow. Good programs handle expected and unexpected errors.

### Common Built-in Exceptions (Table 1-17 — selected)

| Exception | When it happens |
|-----------|-----------------|
| `ZeroDivisionError` | Dividing by zero |
| `TypeError` | Wrong type for an operation |
| `ValueError` | Valid type but invalid value |
| `NameError` | Variable name not defined |
| `IndexError` | List index out of range |
| `KeyError` | Dictionary key not found |
| `IOError` | Input/output failure (file issues) |
| `SyntaxError` | Invalid Python syntax |
| `IndentationError` | Incorrect indentation |

### try / except / finally

- **try** — code that might fail
- **except** — handle specific errors
- **finally** — always runs (cleanup)

**Listing 1-52 — Basic try/except:**


````python
# Listing 1-52: Exception Handling
import sys

try:
    a = 1
    b = 1
    print("Result of a/b:", a / b)

except ZeroDivisionError:
    print("Can't divide by zero")

except TypeError:
    print("Wrong data type, division allowed on numeric types only")

except:
    print("Unexpected error occurred")
````

**Listing 1-53 — Exception handling with file I/O and finally:**


````python
# Listing 1-53: Exception Handling with File Operations
import sys

f = None
try:
    f = open('vehicles.txt', 'r')
    s = f.readline()
    print(s)
    i = int(s.strip())

except IOError as e:
    print("I/O error({0}): {1}".format(e.errno, e.strerror))

except ValueError:
    print("Could not convert data to an integer.")

except:
    print("Unexpected error occurred")

finally:
    if f is not None:
        f.close()
        print("file has been closed")
````

### Figure 1-2: Code Flow for Error Handler


````python
# Figure 1-2: Code flow for error handler
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')
ax.set_title('Figure 1-2: Code Flow for Error Handler', fontsize=14, fontweight='bold')

def box(x, y, w, h, text, color='#e3f2fd'):
    rect = FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.05", facecolor=color, edgecolor='#1565c0', linewidth=2)
    ax.add_patch(rect)
    ax.text(x + w / 2, y + h / 2, text, ha='center', va='center', fontsize=9)

def arrow(x1, y1, x2, y2, label=''):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1), arrowprops=dict(arrowstyle='->', color='#333', lw=1.5))
    if label:
        ax.text((x1 + x2) / 2, (y1 + y2) / 2 + 0.15, label, ha='center', fontsize=8, color='#c62828')

box(3.5, 8.5, 3, 0.8, 'Start: try block', '#c8e6c9')
box(3, 7, 4, 0.8, 'Execute main code', '#e3f2fd')
box(2.5, 5.5, 5, 0.8, 'Exception raised?', '#fff9c4')
box(0.5, 4, 3.5, 0.9, 'except:\nHandle error', '#ffcdd2')
box(6, 4, 3.5, 0.9, 'No error:\nContinue', '#c8e6c9')
box(3, 2.2, 4, 0.8, 'finally:\nCleanup (always runs)', '#e1bee7')
box(3.5, 0.8, 3, 0.8, 'Exit program', '#cfd8dc')

arrow(5, 8.5, 5, 7.8)
arrow(5, 7, 5, 6.3)
arrow(3.5, 5.5, 2.25, 4.9, 'Yes')
arrow(6.5, 5.5, 7.75, 4.9, 'No')
arrow(2.25, 4, 4, 3)
arrow(7.75, 4, 6, 3)
arrow(5, 2.2, 5, 1.6)
plt.tight_layout()
plt.show()
````
